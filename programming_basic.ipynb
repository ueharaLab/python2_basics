### 1. 整数の四則演算

```python
a = 1 + 3
print(a)
b = 3*4
print(b)
c = 5/4
print(c)
d= 5//4
print(d)
e = 5%3
print(e)
```

### 3. 演算の優先度
#### 1) 数学と同じ規則。足し算よりも掛け算、掛け算よりべき乗が優先 etc  
#### 2) 整数と実数では実数が優先

```python
a=2+3*2**2
print(a)
b=(2+3)*2**2
print(b)
c=((2+3)*2)**2
print(c)
d=1.0 + 2
print(d)
```
